```python
import matplotlib.pyplot as plt
import pandas as pd
seattle = pd.read_csv('seattle_weather.csv')
austin = pd.read_csv('austin_weather.csv')
```

```python
seattle1 = seattle[seattle['STATION'] == 'USC00456295']
austin1 = austin[austin['STATION'] == 'USW00013904']
fig, ax = plt.subplots() # figure object is a container
ax.plot(seattle1['DATE'], seattle1['MLY-PRCP-NORMAL'], marker = 'x', linestyle = '--', color='b')
ax.plot(austin1['DATE'], austin1['MLY-PRCP-NORMAL'], marker = 'o', linestyle='None', color='r')
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
ax.set_xticks(range(1, 13))
ax.set_xticklabels(months)
ax.set_xlabel('Time (months)')
ax.set_ylabel('Precipitation (inches)')
ax.set_title('Weather in Seattle and Austin')
plt.show()
```

```python
#Small multiples
fig, ax = plt.subplots(2, 1, sharey=True) #ax.shape (3, 2) AND normalize the y-axis
ax[0].plot(seattle1['DATE'], seattle1['MLY-PRCP-NORMAL'], color='b')
ax[0].plot(seattle1['DATE'], seattle1['MLY-PRCP-25PCTL'], linestyle = '--', color='b')
ax[0].plot(seattle1['DATE'], seattle1['MLY-PRCP-75PCTL'], linestyle = '--', color='b')
ax[1].plot(austin1['DATE'], austin1['MLY-PRCP-NORMAL'], color='r')
ax[1].plot(austin1['DATE'], austin1['MLY-PRCP-25PCTL'], linestyle = '--', color='r')
ax[1].plot(austin1['DATE'], austin1['MLY-PRCP-75PCTL'], linestyle = '--', color='r')
ax[1].set_xlabel('Time (months)')
ax[0].set_ylabel('Precipitation (inches)')
ax[1].set_ylabel('Precipitation (inches)')
ax[0].set_xticks(range(1, 13))
ax[0].set_xticklabels(months)
ax[1].set_xticks(range(1, 13))
ax[1].set_xticklabels(months)
ax[0].set_title('Weather in Seattle and Austin')
plt.show()
```

```python
fig, ax = plt.subplots(2, 2)
ax[0, 0].plot(seattle1['DATE'], seattle1['MLY-PRCP-NORMAL'])
ax[0, 1].plot(seattle1['DATE'], seattle1['MLY-TAVG-NORMAL'])
ax[1, 0].plot(austin1['DATE'], austin1['MLY-PRCP-NORMAL'])
ax[1, 1].plot(austin1['DATE'], austin1['MLY-TAVG-NORMAL'])
plt.show()
```

```python
climate = pd.read_csv('climate_change.csv', index_col='date', parse_dates=['date'])
fig, ax = plt.subplots()
ax.plot(climate.index, climate['co2'], label='CO2 Levels')
ax.set_xlabel('Time')
ax.set_ylabel('CO2 (ppm)')
ax.set_title('CO2 Levels Over Time')
ax.legend()
plt.show()
```

```python
sixties = climate["1960-01-01":"1969-12-31"]
sixties.head()
plt.plot(sixties.index, sixties['co2'])
plt.show()
```

```python
#Twin Axes
fig, ax = plt.subplots()
ax.plot(climate.index, climate['co2'], color = 'r')
ax.set_xlabel("Time")
ax.set_ylabel('CO2 (ppm)', color='r')
ax.tick_params('y', colors='r')
ax2 = ax.twinx()
ax2.plot(climate.index, climate['relative_temp'], color = 'b')
ax2.set_ylabel('Relative Temperature', color = 'b')
ax2.tick_params('y', colors='b')
plt.show()
```

```python
def plot_timeseries(axes, x, y, color, xlabel, ylabel):
    axes.plot(x, y, color=color)
    axes.set_xlabel(xlabel)
    axes.set_ylabel(ylabel, color = color)
    axes.tick_params('y', colors=color)

fig, ax = plt.subplots()
plot_timeseries(ax, climate.index, climate['co2'], 'blue', 'Time', 'CO2 (ppm)')
ax2 = ax.twinx()
plot_timeseries(ax2, climate.index, climate['relative_temp'], 'red', 'Time', 'Relative temperature (Celcius)')
plt.show()
```

```python
#Annotations
def plot_timeseries(axes, x, y, color, xlabel, ylabel):
    axes.plot(x, y, color=color)
    axes.set_xlabel(xlabel)
    axes.set_ylabel(ylabel, color = color)
    axes.tick_params('y', colors=color)

fig, ax = plt.subplots()
plot_timeseries(ax, climate.index, climate['co2'], 'blue', 'Time', 'CO2 (ppm)')
ax2 = ax.twinx()
plot_timeseries(ax2, climate.index, climate['relative_temp'], 'red', 'Time', 'Relative temperature (Celcius)')
ax2.annotate('>1 degree', xy=(pd.Timestamp('2015-10-06'), 1), xytext=(pd.Timestamp('2008-10-06'), -0.2), arrowprops={'arrowstyle': '->', "color":'gray'})
plt.show()
```
