```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
height = [171, 162, 153, 204, 195, 186, 177, 168, 159]
weight = [120, 136, 148, 175, 137, 165, 200, 149, 140]
gender = ['Female', 'Female', 'Female', 'Female', 'Male', 'Male', 'Male', 'Male', 'Male']
sns.scatterplot(x = height, y = weight)
plt.show()
sns.countplot(x = gender)
plt.show()
```

```python
mpg = pd.read_csv('mpg.csv')
countries = pd.read_csv('countries-of-the-world.csv')
students = pd.read_csv('student-alcohol-consumption.csv')
young = pd.read_csv('young-people-survey-responses.csv', index_col=0)
sns.countplot(y='Region', data = countries)
plt.show()
```

```python
palette_colors = {'Rural' : 'Green', 'Urban' : 'Blue'}
sns.countplot(x = 'school', hue='location', hue_order = ['Rural', 'Urban'], palette=palette_colors, data = students)
plt.show()
```

```python
#relplots
sns.relplot(x="G1", y="G3",  data = students, kind="scatter", col="schoolsup", col_order=["yes", "no"], row = 'famsup', row_order = ['yes', 'no'])
plt.show()
```

```python
# relationships plot
sns.set_palette(None)
sns.relplot(x="horsepower", y="mpg", data=mpg, kind="scatter", size="cylinders", hue = 'cylinders')
plt.show()
```

```python
# Create a scatter plot of acceleration vs. mpg
sns.relplot(x = 'acceleration', y = 'mpg', data = mpg, hue = 'origin', kind = 'scatter', style = 'origin')
plt.show()
```

```python
#line plots
# Add markers and make each line have the same style
sns.relplot(x="model_year", y="horsepower", data=mpg, kind="line", errorbar = None, style="origin", hue="origin", dashes = False, markers = True)
plt.show()
```

```python
#Categorical plots: Countplot and Barplot
# Separate into column subplots based on age category
sns.catplot(y="Internet usage", data = young, kind="count", col = 'Gender')
plt.show()

sns.catplot(kind = 'bar', data = young, y = 'Mathematics', x = 'Gender')
plt.show()
```

```python
# List of categories from lowest to highest
category_order = ["<2 hours", "2 to 5 hours", "5 to 10 hours", ">10 hours"]
sns.catplot(x="study_time", y="G3", data=students, kind="bar", order=category_order, errorbar = None)
plt.show()
```

```python
#Boxplot
# Create a box plot with subgroups and omit the outliers
sns.catplot(kind = 'box', data = students, x = 'internet', y = 'G3', hue = 'location', whis = 2.0)
# Show plot
plt.show()
```

```python
#Point plot
from numpy import median
# Plot the median number of absences instead of the mean
sns.set_style('darkgrid')
sns.set_palette('RdBu_r')
sns.set_context('talk')
sns.catplot(x="romantic", y="absences", data=students, kind="point", errorbar = None, hue="school", estimator = median)
plt.show()
```

```python
# Set the style to "darkgrid"
sns.set_context('paper')
sns.set_style('darkgrid')
# Set a custom color palette
palette = ["#39A7D0", "#36ADA4"]
sns.set_palette(palette)
# Create the box plot of age distribution by gender
g = sns.catplot(x="Gender", y="Age", data=young, kind="box")
g.figure.suptitle('Gender vs. Age', y = 1.02)
# Show plot
plt.show()
```

```python
#Titles and Labels 
#FacetGrid vs AxesSubplot
g = sns.catplot(data = countries, kind = 'box', x = 'Region', y = 'Birthrate')
plt.xticks(rotation=90)
g.set(xlabel = 'Regions', ylabel = 'Birthrate (mln)')
```

```python
# Set the figure style to "dark"
sns.set_style("dark")
g = sns.catplot(x="Village - town", y="Techno", data=young, kind="bar", col = 'Gender')
g.figure.suptitle("Percentage of Young People Who Like Techno", y=1.02)
g.set(xlabel="Location of Residence", ylabel="% Who Like Techno")
plt.show()
```
