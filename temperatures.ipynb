```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_csv('temperatures.csv')
```

```python
df.set_index('city').loc['Abidjan':'Addis Abeba'].sort_index(level='Country', ascending=False).reset_index(drop=False)
```

```python
df.loc[:, 'country':].drop_duplicates()
```

```python
df.set_index('date').sort_index().loc['2010':'2011']
```

```python
df.set_index(['city','country']).sort_index().loc[('Abidjan','Côte D\'Ivoire') : ('Ahmadabad','India')]#['avg_temp_c'].agg(['min', 'max', 'mean', 'median'])
```

```python
print(df.set_index(['country', 'city']).sort_index().loc[('India', 'Hyderabad'):('Iraq', 'Baghdad'), 'date':'avg_temp_c'])
```

```python
df.pivot_table(values='avg_temp_c', index='country', sort=True, aggfunc=['min', 'max', 'mean', 'median'])
```

```python
df.groupby(['country', 'city'])['avg_temp_c'].agg(['min', 'max', 'mean', 'median'])
```

```python
df['avg_temp_c'].hist()
plt.show()
```

```python
df.groupby('country')['avg_temp_c'].agg(['min', 'max', 'mean', 'median']).plot()
```

```python
df.plot(x='country', y='avg_temp_c')
```

```python
rus = df[df['country'] == 'Russia'].set_index('date').sort_index().reset_index()
rus.plot(x = 'date', y = 'avg_temp_c', xlabel='Date', ylabel='Temperature', kind='line')
plt.show()
```

```python
df[df['country'] == 'India']['avg_temp_c'].hist(alpha=0.8)
df[df['country'] == 'Russia']['avg_temp_c'].hist(alpha=0.8)
plt.legend(['Russia', 'India'])
plt.show()
```
