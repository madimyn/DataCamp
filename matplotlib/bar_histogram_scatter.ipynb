```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
medals = pd.read_csv('medals_by_country_2016.csv', index_col=0)
fig, ax = plt.subplots()
ax.bar(medals.index, medals['Gold'], label = 'Gold')
ax.bar(medals.index, medals['Silver'], bottom = medals['Gold'], label = 'Silver')
ax.bar(medals.index, medals['Bronze'], bottom = medals['Gold'] + medals['Silver'], label = 'Bronze')
ax.tick_params(axis='x', rotation=90)
ax.set_xlabel('Number of medals')
ax.legend()
plt.show()
```

```python
summer = pd.read_csv('summer2016.csv', index_col=0)
rowing = summer[(summer['Sex'] == 'M') & (summer['Sport'] == 'Rowing')]
gymnastics = summer[(summer['Sex'] == 'M') & (summer['Sport'] == 'Gymnastics')]
fig, ax = plt.subplots()
ax.bar('Rowing', rowing['Height'].mean())
ax.bar('Gymnastics', gymnastics['Height'].mean())
ax.set_xlabel('Height (cm)')
plt.show()
```

```python
#histogram
fig, ax = plt.subplots()
ax.hist(rowing['Height'], label='Rowing', bins = [150, 160, 170, 180, 190, 200, 210], histtype='step')
ax.hist(gymnastics['Height'], label='Gymnastics', bins = [150, 160, 170, 180, 190, 200, 210], histtype='step')
ax.set_ylabel('Number of observations')
ax.set_xlabel('Height (cm)')
ax.legend()
plt.show()
```

```python
#Error bars in barchart
fig, ax = plt.subplots()
ax.bar('Rowing', rowing['Height'].mean(), yerr=rowing['Height'].std(),label='Rowing')
ax.bar('Gymnastics', gymnastics['Height'].mean(), yerr = gymnastics['Height'].std(), label='Gymnastics')
ax.set_ylabel('Height (cm)')
ax.legend()
plt.show()
```

```python
seattle = pd.read_csv('seattle_weather.csv', index_col=0)
austin = pd.read_csv('austin_weather.csv', index_col=0)
seattle = seattle[seattle.index == 'USC00456295']
fig, ax = plt.subplots()
ax.errorbar(seattle['DATE'], seattle['MLY-TAVG-NORMAL'], yerr = seattle['MLY-TAVG-STDDEV'], label = 'Seattle')
ax.errorbar(austin['DATE'], austin['MLY-TAVG-NORMAL'], yerr = austin['MLY-TAVG-STDDEV'], label = 'Austin')
ax.legend()
ax.set_ylabel('Time (months)')
ax.set_xlabel('Temperature (F)')
plt.show()
```

```python
#boxplots
fig, ax = plt.subplots()
ax.boxplot([rowing['Height'], gymnastics['Height']])
ax.set_xticklabels(['Rowing', 'Gymnastics'])
ax.set_ylabel('Height (cm)') #3 outliers for rowers and 1 for gymnasts
plt.show()
```

```python
#Scatterplot
climate_change = pd.read_csv('climate_change.csv', index_col=['date'], parse_dates=['date'])
fig, ax = plt.subplots()
ax.scatter(climate_change['co2'], climate_change['relative_temp'])
ax.set_ylabel('Relative temperature (Celcius)')
ax.set_xlabel('CO2 (ppm)')
plt.show()
```

```python
eighties = climate_change['1980-01-01':'1989-12-31']
nineties = climate_change['1990-01-01':'1999-12-31']
fig, ax = plt.subplots()
ax.scatter(eighties['co2'], eighties['relative_temp'], color = 'r', label='80s')
ax.scatter(nineties['co2'], nineties['relative_temp'], color = 'b', label='90s')
ax.set_ylabel('Relative temperature (Celcius)')
ax.set_xlabel('CO2 (ppm)')
ax.legend()
plt.show()

fig, ax = plt.subplots()
ax.scatter(climate_change['co2'], climate_change['relative_temp'], c = climate_change.index)
plt.show()
```

```python
plt.style.use('default') #default, bmh, Solarize_Light2, ggplot
fig, ax = plt.subplots()
ax.plot(seattle['DATE'], seattle['MLY-TAVG-NORMAL'], label = 'Seattle')
ax.plot(austin['DATE'], austin['MLY-TAVG-NORMAL'], label = 'Austin')
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
ax.set_xticks(range(1, 13))
ax.set_xticklabels(months)
ax.set_ylabel('Time (months)')
ax.set_xlabel('Temperature (F)')
ax.legend()
plt.show()
```

```python
sports = summer['Sport'].unique()
fig, ax = plt.subplots()
for sport in sports:
    sport_df = summer[(summer['Sport'] == sport)]
    ax.bar(sport, sport_df['Weight'].mean(), yerr = sport_df['Weight'].std())
ax.set_ylabel('Weight (kg)')
ax.tick_params(axis='x', rotation=90)
plt.show()
```
